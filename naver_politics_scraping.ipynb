```python
import urllib.request
from bs4 import BeautifulSoup
import webbrowser
from selenium import webdriver
import time
import re
from konlpy.tag import Okt
from collections import Counter
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
from wordcloud import WordCloud
import stylecloud
```

```python
options = webdriver.ChromeOptions()
# options.add_argument('headless')
options.add_argument('window-size=1920x1080')
# options.add_argument("disable-gpu")
# # 혹은 options.add_argument("--disable-gpu")
wd = webdriver.Chrome('/home/ai106/PycharmProjects/pythonProject/NLP/chromedriver', options=options)

naver_url = "https://search.naver.com/search.naver?where=news&sm=tab_pge&query=%EC%A0%95%EC%B9%98&sort=0&photo=0&field=0&pd=3&ds=2021.06.01&de=2021.06.08&cluster_rank=29&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so:r,p:from20210601to20210608,a:all&start="

titles_str = ''
for i in range(1, 3992, 10):
    print(i)
    full_url = naver_url + str(i)
    wd.get(full_url)
    time.sleep(0.3)

    html = wd.page_source
    soup_nv = BeautifulSoup(html, 'html.parser')
    articles = soup_nv.find_all('a', class_='news_tit')
    for title_area in articles:
        title = title_area.attrs['title']
        titles_str += ' '+title
```

```python
titles_str
```

```python
titles_str = re.sub(r'[^\w]', ' ', titles_str)
titles_str
```

```python
with open ("titles_str.txt", "w") as f:
    f.write(titles_str)
```

```python
nlp = Okt()
titles_N = nlp.nouns(titles_str)
titles_N
```

```python
count = Counter(titles_N)
count
```

```python
word_count = dict()
for tag, counts in count.most_common(300):
    if(len(str(tag))>1):
        word_count[tag] = counts
        print("%s : %d" % (tag, counts))
```

```python
del(word_count['정치'])
word_count
```

```python
font_path = "/usr/share/fonts/truetype/nanum/NanumGothicBold.ttf"
font_name = font_manager.FontProperties(fname=font_path).get_name()
matplotlib.rc('font', family=font_name)
```

```python
plt.figure(figsize=(20,8))
plt.xlabel('키워드')
plt.ylabel('빈도수')
plt.grid(True)
sorted_Keys = sorted(word_count, key=word_count.get, reverse=True)
sorted_Values = sorted(word_count.values(), reverse=True)
plt.bar(range(len(word_count)), sorted_Values, align='center')
plt.xticks(range(len(word_count)), list(sorted_Keys), rotation='75')
plt.show()
```

```python
wc = WordCloud(font_path, background_color='ivory', width=800, height=600)
cloud = wc.generate_from_frequencies(word_count)
plt.figure(figsize=(12,12))
plt.imshow(cloud)
plt.axis('off')
plt.show()
```
